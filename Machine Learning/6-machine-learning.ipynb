**Categorical variables** are 2 types.
1. **Nominal**
2. **Ordinal**

Nominal categorical variables are **does not follow order** but Ordinal categorical variables **follows order.**

Ex for Nominal : MALE, FEMALE

Ex for Ordinal : phd, master,degree

```python
# Import pandas library
import pandas as pd
 
# initialize list of lists
data = [
    ['Software engineer',20,550000], ['Doctor',20,550000], ['Actor',20,550000],
    ['Software engineer',22,590000], ['Doctor',23,650000], ['Actor',22,370000],
    ['Software engineer',24,650000], ['Doctor',25,750000], ['Actor',23,780000],
    ['Software engineer',29,750000], ['Doctor',29,950000], ['Actor',24,910000]
]
 
# Create the pandas DataFrame
df = pd.DataFrame(data, columns = ['designation','age','salary'])
 
# print dataframe.
df
```

designation column is a categorical column , we need to convert to munbers using 
1. pandas get dummies (or)
2. One Hot Encoder

# 1. pandas get dummies

```python
import pandas as pd
dummies = pd.get_dummies(df['designation'])
dummies
```

```python
merged = pd.concat([df,dummies], axis='columns')
merged
```

drop categorical column and any on of the dummy column (bcz with other dummy columns we can get the info of this column so drop it)

```python
final  = merged.drop(['designation','Actor'],axis='columns')
final
```

# 2. One Hot Encoder

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder

# creating instance of one-hot-encoder
enc = OneHotEncoder(handle_unknown='ignore')

# passing bridge-types-cat column (label encoded values of bridge_types)
enc_df = pd.DataFrame(enc.fit_transform(df[['designation']]).toarray())
enc_df
```

```python

# merge with main df bridge_df on key values
bridge_df = df.join(enc_df)
bridge_df = bridge_df.drop(['designation',0],axis='columns') 
bridge_df
```
