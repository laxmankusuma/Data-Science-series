**Logistic Regression** is used to solve classification problem. 

If the outcome is only 2 then we call **binary logistic regression**, if the out come is more than 2 then we call it as **multiclass logistic regression**.

Logistic regression uses **sigmoid(logit) Function**

sigmoid function convert input in the range of 0 to 1.

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Import pandas library
import pandas as pd
 
# initialize list of lists
data = [[9,0],[10,0],[11,0],[12,0],[13,0],[14,0],[15,0],[16,0],[20,1],[21,1],[22,1],[23,1],[24,1],[25,1],[26,1],[27,1]]
 
# Create the pandas DataFrame
df = pd.DataFrame(data, columns = ['age','insurance_taking'])
 
# print dataframe.
df
```

```python
plt.scatter(df['age'], df['insurance_taking'], marker='+',color='red')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['age']], df['insurance_taking'], test_size=0.2, random_state=42)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
model.score(X_test,y_test)
```

```python
model.predict_log_proba(X_test).astype('int64') 
```
